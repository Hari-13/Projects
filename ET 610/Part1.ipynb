```python

sill_score_memory= sill_score.iloc[:,3:11]
sill_score_cognitive= sill_score.iloc[:,11:24]
sill_score_compensation=sill_score.iloc[:,24:30]
sill_score_metacognition=sill_score.iloc[:,30:39]
sill_score_affective=sill_score.iloc[:,39:45]
sill_score_social=sill_score.iloc[:,45:51]
```

```python

sill_score['memory']=sill_score_memory.sum(axis=1)/8
sill_score['cognitive']=sill_score_cognitive.sum(axis=1)/13
sill_score['compensation']=sill_score_compensation.sum(axis=1)/6
sill_score['metacognition']=sill_score_metacognition.sum(axis=1)/9
sill_score['affective']=sill_score_affective.sum(axis=1)/6
sill_score['social']=sill_score_social.sum(axis=1)/6
```
